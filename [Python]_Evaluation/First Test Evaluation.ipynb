```python
for s in data:
    for t in data[s]:
        plt.figure()
        plt.title(s + " Tile size: " + t)
        for dat in data[s][t]:
            d = data[s][t][dat]
            (d['tiles shaded']*d['tile size']*d['tile size']/1000000).plot()

        plt.legend(list(data[s][t].keys()),loc='lower right',ncol=2, prop={'size': 8})
        plt.xlabel("Frame Number")
        plt.ylabel("Pixels Shaded (in millions)")
        plt.savefig(s+t+".pdf")
```

```python
def toPer(i):
    return "{0:.0f}%".format(i *100)
```

```python
tab = []
for t in data[s]:
    for dat in data[s][t]:
        d = data["sponza"][t][dat]
        m1 = (d['tiles shaded']*d['tile size']*d['tile size']/(d['screen width']*d['screen height']))
        d = data['lostEmpire'][t][dat]
        m2 = (d['tiles shaded']*d['tile size']*d['tile size']/(d['screen width']*d['screen height']))
        tab += [[t,dat,toPer(m1.median()),toPer(np.var(m1)),toPer(m2.median()),toPer(np.var(m2))]]
        
tab = pd.DataFrame(tab, columns = ['tile size','screen size', 'sponza OSS/Standard', 'sponza standard deviation', 'lost empire OSS/Standard', 'lost empire standard deviation'])

print(tab.to_latex())
```
