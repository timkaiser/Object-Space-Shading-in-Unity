```python
#imports
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
#read data
labels = ['tile size','screen size','cam id 1','tiles cam 1','cam id 2','tiles cam 2', 'common tiles']
tilesizes = [8,16,32,64]
screensizes = [' 800x600',' 1024x768',' 1920x1080',' 3840x2160']
dataDict = {
    "sponza" : pd.read_csv("SecondTest\SecondTestSponza.csv",names =  labels),
    "lostEmpire" : pd.read_csv("SecondTest\SecondTestLostEmpire.csv",names =  labels),
}
```

```python
def toPer(i):
    return "{0:.0f}%".format(i *100)
```

```python
#add percentage data
for data in dataDict.values():
    data['% cam 1'] = data['common tiles']/data['tiles cam 1']
    data['% cam 2'] = data['common tiles']/data['tiles cam 2']
```

```python
#group data
groupedData = data[['tile size','screen size','tiles cam 1','tiles cam 2', 'common tiles','% cam 1','% cam 2']].groupby(['tile size','screen size']).mean()
```

```python
df = pd.DataFrame([],columns=['tile size','1&2','1&3','1&4','1&5','2&3','2&4','2&5','3&4','3&5','4&5'])
i = 0
scene = "sponza"
for t in tilesizes:
    for s in screensizes:
        row = [t]
        for c1 in range(0,4):
            for c2 in range(c1+1,5):
                data = dataDict[scene]
                row += [data[(data["tile size"] == t) 
                        & (data["screen size"] == s) 
                        & (data["cam id 1"] == c1)
                        & (data["cam id 2"] == c2)]["% cam 1"].values[0]]
        df.loc[i] = row
        i += 1
    
print(df.to_latex())
```

```python
df2 = df.groupby(['tile size']).mean()

for old_col in df2:
    new_col = [] 
    for val in df2[old_col]:
        new_col += [toPer(val)]
    df2[old_col] = new_col

print(df2.to_latex())
               
```
